### Carregue os dados e encontre as medidas de dispersão e de shape (IQR, kutorsis e skewness) para a coluna assault e fraud.

```python
import pandas as pd

df = pd.read_excel("dataset/USArrests.xlsx")
df.head(5)
```

```python
from scipy.stats import iqr

print(f"IQR para Assault: {iqr(df['Assault'])}")
print(f"IQR para Fraud: {iqr(df['Fraud'])}")
```

```python
from scipy.stats import kurtosis

print(f"kurtosis para Assault: {kurtosis(df['Assault'], fisher=True)}")
print(f"kurtosis para Fraud: {kurtosis(df['Fraud'], fisher=True)}")
```

```python
from scipy.stats import skew

print(f"Skewness para Assault: {skew(df['Assault'])}")
print(f"Skewness para Fraud: {skew(df['Fraud'])}")
```

### Obter o zscore para a coluna assault.

```python
from scipy.stats import zscore

print(f"Z-score para Assault: {zscore(df['Assault'])}")
```

### Detecte outliers da coluna urbanpop usando o método de IQR.

```python
import numpy as np
def outlier_treat(column: str):
    q3 = np.quantile(df[column], 0.75)
    q1 = np.quantile(df[column], 0.25)

    iqr_manual = q3-q1
    upper_limit = q3 + 1.5 * iqr_manual
    lower_limit = q1 - 1.5 * iqr_manual

    return df[(df[column] > lower_limit) & (df[column] < upper_limit)]
```

```python
outlier_treat("UrbanPop")
```

```python
iqr(df["UrbanPop"])
```

### Detecte outliers da coluna urbanpop usando o método de zscore.

```python
from scipy.stats import zscore

np.abs(zscore(df["UrbanPop"]))
df[np.abs(zscore(df["UrbanPop"])) < 3]
```

### Calcule o range, o desvio padrão e variância nas colunas urbanpop e assault.

```python
print(f"Range para Assault: {df['Assault'].max() - df['Assault'].min()}")
print(f"Range para UrbanPop: {df['UrbanPop'].max() - df['UrbanPop'].min()}")
```

```python
print(f"Std para Assault: {df['Assault'].std()}")
print(f"Std para UrbanPop {df['UrbanPop'].std()}")
```

```python
print(f"Var para Assault: {df['Assault'].var()}")
print(f"Var para UrbanPop {df['UrbanPop'].var()}")
```
