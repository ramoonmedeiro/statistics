```python
import numpy as np
import seaborn as sns
import pandas as pd
from scipy import stats
```

```python
# criando dados aleatórios mas que seguem a distribuição normal
dados_normal = stats.norm.rvs(size=1000)
```

```python
dados_normal
```

```python
dados_normal.mean(), dados_normal.std()
```

```python
sns.histplot(dados_normal, kde=True)
```

# Teorema central do limite

```python
import numpy as np
alturas = np.random.randint(150, 190, 500)
alturas.mean()
```

```python
import seaborn as sns
sns.histplot(alturas, kde=True);
```

```python
medias = [np.random.randint(150, 190, 500).mean() for _ in range(1000)]
```

```python
sns.histplot(medias, kde=True);
```

# Distribuição Gama

```python
from scipy.stats import gamma
```

```python
dados_gama = gamma.rvs(a=1, size=1000)
# a = parâmetro de decaimento
```

```python
dados_gama.mean()
```

```python
sns.histplot(dados_gama, kde=True);
```

# Distribuição expoencial

```python
from scipy.stats import expon
dados_expon = expon.rvs(size=1000)
```

```python
sns.histplot(dados_expon, kde=True);
```

# Distribuição de Bernoulli

```python
from scipy.stats import bernoulli

dados_bernoulli = bernoulli.rvs(size=1000, p=0.5)
```

```python
dados_bernoulli
```

```python
sns.histplot(dados_bernoulli, kde=True);
```

# Distribuição Binomial

```python
from scipy.stats import binom
dados_binom = binom.rvs(size=1000, n=10, p=0.5)
```

```python
sns.histplot(dados_binom, kde=True);
```

# Distribuição de Poisson

```python
from scipy.stats import poisson
dados_poisson = poisson.rvs(size=1000, mu=3)
```

```python
sns.histplot(dados_poisson, kde=True);
```
